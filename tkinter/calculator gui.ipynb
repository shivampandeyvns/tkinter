```python
from tkinter import *
```

```python
root=Tk()
e=Entry(root,borderwidth=3,width=63)
e.grid(row=0,column=1,columnspan=4)
e1=Entry(root,borderwidth=3)
e1.grid(row=0,column=0)
def add():
    first_number=e.get()
    global f_num
    global math
    math='addition'
    f_num=int(first_number)
    e.delete(0, END)
    e1.insert(0,str(f_num)+'+')
    #result=Label(root, text=int(num1.get())+int(num2.get()))
    #result.grid(row=6,column=1)
def diff():
    first_number=e.get()
    global f_num
    global math
    math='difference'
    f_num=int(first_number)
    e1.insert(0,str(f_num)+'-')
    e.delete(0, END)
def mul():
    first_number=e.get()
    global f_num
    global math
    math='multiplication'
    f_num=int(first_number)
    e1.insert(0,str(f_num)+'*')
    e.delete(0, END)
def div():
    first_number=e.get()
    global f_num
    global math
    math='division'
    f_num=int(first_number)
    e1.insert(0,str(f_num)+'/')
    e.delete(0, END)
def button_clear():
    e.delete(0, END)
def button_click(number):
    current=e.get()
    e.delete(0, END)
    e.insert(0,str(current) + str(number))
def button_equal():
    second_number=e.get()
    e.delete(0, END)
    if(math=='addition'):
        e.insert(0, f_num+int(second_number))
        e1.delete(0, END)
        e1.insert(0,str(f_num)+'+'+str(second_number)+'=')
    elif(math=='difference'):
        e.insert(0, f_num-int(second_number))
        e1.delete(0, END)
        e1.insert(0,str(f_num)+'-'+str(second_number)+'=')
    elif(math=='multiplication'):
        e.insert(0, f_num*int(second_number))
        e1.delete(0, END)
        e1.insert(0,str(f_num)+'*'+str(second_number)+'=')
    elif(math=='division'):
        e.insert(0, f_num/int(second_number))
        e1.delete(0, END)
        e1.insert(0,str(f_num)+'/'+str(second_number+'='))
button_1=Button(root,text='1',padx=40,pady=20,command=lambda: button_click(1))
button_2=Button(root,text='2',padx=40,pady=20,command=lambda: button_click(2))  
button_3=Button(root,text='3',padx=40,pady=20,command=lambda: button_click(3))  
button_4=Button(root,text='4',padx=40,pady=20,command=lambda: button_click(4))  
button_5=Button(root,text='5',padx=40,pady=20,command=lambda: button_click(5))  
button_6=Button(root,text='6',padx=40,pady=20,command=lambda: button_click(6))  
button_7=Button(root,text='7',padx=40,pady=20,command=lambda: button_click(7))  
button_8=Button(root,text='8',padx=40,pady=20,command=lambda: button_click(8))  
button_9=Button(root,text='9',padx=40,pady=20,command=lambda: button_click(9))  
button_0=Button(root,text='0',padx=40,pady=20,command=lambda: button_click(0))  
sum_buttom=Button(root, text='+',padx=40,pady=20,command=add)
diff_buttom=Button(root, text='-',padx=40,pady=20,command=diff)
mul_buttom=Button(root, text='*',padx=40,pady=20,command=mul)
div_buttom=Button(root, text='/',padx=40,pady=20,command=div)
clear_button=Button(root,text='CLEAR',padx=25.5,pady=20,command=button_clear)
equal_button=Button(root,text='=',padx=40,pady=20,command=button_equal)
button_1.grid(row=3,column=2)
button_2.grid(row=3,column=1)
button_3.grid(row=3,column=0)
button_4.grid(row=2,column=2)
button_5.grid(row=2,column=1)
button_6.grid(row=2,column=0)
button_7.grid(row=1,column=2)
button_8.grid(row=1,column=1)
button_9.grid(row=1,column=0)
button_0.grid(row=4,column=1)
clear_button.grid(row=4,column=0)
equal_button.grid(row=4,column=2)
sum_buttom.grid(row=1,column=3)
diff_buttom.grid(row=2,column=3)
mul_buttom.grid(row=3,column=3)
div_buttom.grid(row=4,column=3)
root.mainloop()
```
